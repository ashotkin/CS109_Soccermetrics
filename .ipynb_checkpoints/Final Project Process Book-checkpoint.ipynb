```python
## This code just goes through and assigns a club to each player for each game.

#This frame (if run) prints out the index every 10000 iterations, so it's possible
#to track how far along the code is. 
fails = []
positions = []

for index, row in player_games.iterrows():
    player,year = player_games.ix[index][['Name','Year']]
    temp = players[players['Player'] == player]
    temp = temp[temp['Year'] == str(year)] 
    club = temp['Club'].tolist()
    if club:
        player_games.loc[index,['Club']] = temp['Club'].tolist()[0]
        positions.append(temp['Position'].tolist()[0])
    else:
        player_games.loc[index,['Club']] = 'failure'
        positions.append('failure')
        fails.append(index)
    if index % 10000 == 0:
        print index
player_games['POS'] = positions
```

```python
## Run this to create the game dictionary
offense = ['F', 'F-M', 'M', 'M-F']
temp = player_games.drop_duplicates()
temp = temp[temp['POS'].isin(offense)]
temp = temp.groupby(by =['Club','Date','Name'], as_index=False).sum()
teams = set(temp['Club'])
games = {}
for team in teams:
    game = {}
    temp2 = temp[temp['Club'] == team]
    dates = set(temp2['Date'])
    for date in dates:
        players = {}
        tempPlayers = temp2[temp2['Date'] == date]
        for player in set(tempPlayers['Name']):
            players[player] = (tempPlayers[tempPlayers['Name'] == player])['Minutes'].tolist()[0]
        game[date] = players
    games[team] = game
```

```python
keepers = player_games.drop_duplicates()
keepers = keepers[keepers['POS'].isin(offense)]
keepers = keepers.groupby(by =['Club','Date','Name'], as_index=False).sum()
teams = set(temp['Club'])
goalies = {}
for team in teams:
    goalie = {}
    keepers2 = keepers[keepers['Club'] == team]
    dates = set(keepers2['Date'])
    for date in dates:
        guys = {}
        tempGoalies = keepers2[keepers2['Date'] == date]
        for player in set(tempGoalies['Name']):
            guys[player] = (tempGoalies[tempGoalies['Name'] == player])['Minutes'].tolist()[0]
        goalie[date] = guys
    goalies[team] = goalie
```

```python
## This code for scraping the MLS data on what players played in each game took an entire day to run and required manual fixes to any player
## that caused an issue with the scraping and so data should in the future be loaded in from the appropriate CSV and not re-scraped.

playerlist = list(set(players["Player"].tolist()))
ignore_list = ["Rafhinha","Lyle Martin","TOR","Aaron Ponce","Marcos Mondaini","Rodolfo Espinoza","Cristian Fernandez","Dario Delgado","Eduardo Fernandez","DAL","Junior Burgos","Tico-Tico","Oscar Recio","Cyprian Hedrick","Santiago Hirsig","Tiyi Shipalane","Gege Soriola","Andrew Quinn","Barry Rice","Syamsir Alam","Mike Holody","J.P. Rodrigues","Steve Purdy","Paulo Araujo Jr.","Erick","Anthony Obodai","Kenny Mansally","Tiago Ulisses","Narciso Fernandes","Richard Marquez","Missael Espinoza","Rafael Marquez","Alexandre Emerson","Maksim Usanov"]
def addPlayerToResults(name):
    j = len(player_games)
    if name not in ignore_list:
        try:
            nameURL = name.lower().replace(' ','-').replace('.','').replace('\'','')
            page = BeautifulSoup(urllib2.urlopen(root_player_URL + nameURL).read(), 'html.parser')
            if not "This player does not currently have a game log." in unicode(page):
                tablediv = [tree for tree in page.findAll("div") if 'class' in tree.attrs.keys() and ("stats-table" in tree.attrs["class"]) and ("Regular Season Game Log" in unicode(tree))][0]
                table = tablediv.findAll("table")[0]
                clubyears = dict()
                for row in table.findAll("tr")[1:]:
                    col = row.findAll("td")
                    date = col[0].string
                    minutes = int(col[4].string)
                    yr = int(date.split('/')[-1])
                    if yr not in clubyears:
                        clubyears[yr] = f(players[players["Player"] == name][players["Year"] == yr]["Club"].tolist())
                    if minutes > 0:
                        player_games.loc[j] = [name,date,clubyears[yr],minutes]
                        j+=1
        except:
            print name
            print playerlist.index(name)
            raise

#[addPlayerToResults(x) for x in playerlist[2494:]]
#player_games.head()
```

```python
##### DON'T RUN

resultpage = "http://www.mlssoccer.com/results/"


results = pd.DataFrame(columns=('Year', 'Club', 'Date', 'Versus', 'Club Score', 'Opposing Score', 'W/L/D'))
n=0
for year in range(2011,2015):
    page = BeautifulSoup(urllib2.urlopen(resultpage + str(year)).read(), "html.parser")
    table = page.find('table')
    for row in table.findAll('tr')[1:]:
        col = row.findAll('td')
        for i in range(len(col)):
            if i == 0:
                team = col[i]
                clubsplit = str(team).split('\n')
                club = clubsplit[2][10:-16]
            else:
                data = str(col[i]).split('>')
                winloss = data[2][0]
                date =  data[3][:-4]
                versussplit = data[4].split()
                versus = ' '
                for j in range(len(versussplit)):
                    versus = versus + ' ' + versussplit[j]
                versus = versus[5:-4]
                split = data[5].split()
                score1 = int(split[0])
                score2 = int(split[2])
                

                if winloss == 'W':
                    if score1<score2:
                        score = score2
                        scoreopp = score1
                    else:
                        score = score1
                        scoreopp = score2
                elif winloss == 'L':
                    if score1<score2:
                        score = score1
                        scoreopp = score2
                    else:
                        score = score2
                        scoreopp = score1
                else:
                    score = score1
                    scoreopp = score2
                    
                results.loc[n] = [year, club, date, versus, score, scoreopp, winloss]
                n+=1

            
results
```
