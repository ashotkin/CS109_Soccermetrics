```python
import urllib2
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# I didn't know how to scrape before. So I used this webpage as a reference:
# http://palewi.re/posts/2008/04/20/python-recipe-grab-a-page-scrape-a-table-download-a-file/

mainpage = "http://www.mlssoccer.com/stats/season"

# the table is continued on other pages
#urls = [mainpage + '?page={}'.format(i) for i in range(1,26)]
#urls.insert(0,mainpage)

playerpage = "http://www.mlssoccer.com/stats/season?"
players = pd.DataFrame(columns=('Player', 'Year', 'Club', 'Position', 'GP', 'GS', 'MINS', 'G', 'A', 'SHTS', 'SOG', 'GWG', 'PKG/A', 'HmG', 'RdG', 'G/90min', 'SC%'))
i=0
for year in range(1996,2015):
    print year
    pageNumber = 0
    last_page = 1
    while pageNumber <= last_page:
        page = BeautifulSoup(urllib2.urlopen(playerpage + "season_year=" + str(year) + "&page=" + str(pageNumber)).read())
        #last_page = int(page.find('li', {"class":"pager-current last"}).contents[0])
        if pageNumber == 0:
            last_page = int(str(page.find('li', {"class":"pager-last last"}).contents[0]).split('=')[3].split('&')[0])
        pageNumber+=1
        table = page.find('table')
        for row in table.findAll('tr')[1:]:
            col = row.findAll('td')
            player = col[0].string
            club = col[1].string
            pos = col[2].string
            gp = int(col[3].string)
            gs = int(col[4].string)
            mins = int(col[5].string)
            g = int(col[6].string)
            a = int(col[7].string)
            shts = int(col[8].string)
            sog = int(col[9].string)
            gwg = int(col[10].string)
            pkga = col[11].string
            hmg = int(col[12].string)
            rdg = int(col[13].string)
            g90min = float(col[14].string)
            sc = float(col[15].string)
            players.loc[i] = [player, year, club, pos, gp, gs, mins, g, a, shts, sog, gwg, pkga, hmg, rdg, g90min, sc]
            i+=1
            
players
```

```python
# I like dataframes, because they're nice to look at, but I'm guessing this may not be the most
# efficient way to store our data.
# If someone else knows a better way to do this, please feel free to fix.


df = pd.DataFrame(columns=('Player', 'Club', 'Position', 'GP')) # going to add the rest of the columns later

i=0 # counter for adding rows to df
for url in urls:
    page = urllib2.urlopen(url).read()
    soup = BeautifulSoup(page)
    table = soup.find('table')
    for row in table.findAll('tr')[1:]:
        col = row.findAll('td')
        player = col[0].string
        club = col[1].string
        pos = col[2].string
        gp = int(col[3].string)
        df.loc[i] = [player, club, pos, gp]
        i = i+1
    
df
```
