## Scraping the data into CSVs

```python
###### NEVER RUN THIS CELL AGAIN ######

# I didn't know how to scrape before. So I used this webpage as a reference:
# http://palewi.re/posts/2008/04/20/python-recipe-grab-a-page-scrape-a-table-download-a-file/

mainpage = "http://www.mlssoccer.com/stats/season"

# the table is continued on other pages
#urls = [mainpage + '?page={}'.format(i) for i in range(1,26)]
#urls.insert(0,mainpage)

playerpage = "http://www.mlssoccer.com/stats/season?"
players = pd.DataFrame(columns=('Player', 'Year', 'Club', 'Position', 'GP', 'GS', 'MINS', 'G', 'A', 'SHTS', 'SOG', 'GWG', 'PKG/A', 'HmG', 'RdG', 'G/90min', 'SC%'))
i=0
for year in range(1996,2015):
    pageNumber = 0
    last_page = 0
    while pageNumber <= last_page:
        page = BeautifulSoup(urllib2.urlopen(playerpage + "season_year=" + str(year) + "&page=" + str(pageNumber)).read(), 'html.parser')
        if pageNumber == 0:
            last_page = int(str(page.find('li', {"class":"pager-last last"}).contents[0]).split('=')[3].split('&')[0])
        pageNumber+=1
        table = page.find('table')
        for row in table.findAll('tr')[1:]:
            col = row.findAll('td')
            player = col[0].string
            club = col[1].string
            pos = col[2].string
            gp = int(col[3].string)
            gs = int(col[4].string)
            mins = int(col[5].string)
            g = int(col[6].string)
            a = int(col[7].string)
            shts = int(col[8].string)
            sog = int(col[9].string)
            gwg = int(col[10].string)
            pkga = col[11].string
            hmg = int(col[12].string)
            rdg = int(col[13].string)
            g90min = float(col[14].string)
            sc = float(col[15].string)
            players.loc[i] = [player, year, club, pos, gp, gs, mins, g, a, shts, sog, gwg, pkga, hmg, rdg, g90min, sc]
            i+=1
            
players
```

```python
##### DON'T RUN THIS ONE EITHER
players.to_csv('players.csv')
```

```python
##### DON'T RUN
teampage = 'http://www.mlssoccer.com/stats/team'

teams = pd.DataFrame(columns=('Club', 'Year', 'Season', 'GP', 'G', 'A', 'SHT', 'SOG', 'FC', 'FS', 'OFF', 'CK', 'PKG', 'PKA'))


for year in range(1996,2015):
    for season in ['REG', 'PS']:
        
        i = len(teams['Club'])
        
        page = BeautifulSoup(urllib2.urlopen(teampage + "?season_year=" + str(year) + "&season_type=" + season).read(), 'html.parser')
        table = page.find('table')
        for row in table.findAll('tr')[1:]:
            col = row.findAll('td')
            club = col[0].string
            gp = int(col[1].string)
            g = int(col[2].string)
            a = int(col[3].string)
            sht = int(col[4].string)
            sog = int(col[5].string)
            fc = int(col[6].string)
            fs = int(col[7].string)
            off = int(col[8].string)
            ck = int(col[9].string)
            pkg = int(col[10].string)
            pka = int(col[11].string)
            
            teams.loc[i] = [club, year, season, gp, g, a, sht, sog, fc, fs, off, ck, pkg, pka]
            i+=1
teams
```

```python
##### DON'T RUN
teams.to_csv('teams.csv')
```

```python
##### DON'T RUN

resultpage = "http://www.mlssoccer.com/results/"


results = pd.DataFrame(columns=('Year', 'Club', 'Date', 'Versus', 'Club Score', 'Opposing Score', 'W/L/D'))
n=0
for year in range(2011,2015):
    page = BeautifulSoup(urllib2.urlopen(resultpage + str(year)).read(), "html.parser")
    table = page.find('table')
    for row in table.findAll('tr')[1:]:
        col = row.findAll('td')
        for i in range(len(col)):
            if i == 0:
                team = col[i]
                clubsplit = str(team).split('\n')
                club = clubsplit[2][10:-16]
            else:
                data = str(col[i]).split('>')
                winloss = data[2][0]
                date =  data[3][:-4]
                versussplit = data[4].split()
                versus = ' '
                for j in range(len(versussplit)):
                    versus = versus + ' ' + versussplit[j]
                versus = versus[5:-4]
                split = data[5].split()
                score1 = int(split[0])
                score2 = int(split[2])
                

                if winloss == 'W':
                    if score1<score2:
                        score = score2
                        scoreopp = score1
                    else:
                        score = score1
                        scoreopp = score2
                elif winloss == 'L':
                    if score1<score2:
                        score = score1
                        scoreopp = score2
                    else:
                        score = score2
                        scoreopp = score1
                else:
                    score = score1
                    scoreopp = score2
                    
                results.loc[n] = [year, club, date, versus, score, scoreopp, winloss]
                n+=1

            
results
```

```python
##### DON'T RUN
results.to_csv('results.csv')
```

```python
###### DON'T RUN
salaries = pd.DataFrame(columns = ('Club', 'Last Name', 'First Name', 'Pos', 'Base Salary', 'Guaranteed Compensation', 'Year'))

for year in range(2007,2015):
    
    n = len(salaries['Club'])
    
    url = 'http://www.hcs.harvard.edu/~hub/' + str(year) + '%20salaries%20by%20player.csv'

    data = pd.read_csv(url)


    for x in range(6,(len(data.columns))):
        del data['Unnamed: ' + str(x)]
 
    data.columns = ['Club', 'Last Name', 'First Name', 'Pos', 'Base Salary', 'Guaranteed Compensation']
    
    years = [year for y in range(len(data['Club']))]
    
    data['Year'] = years
    
    for z in range(len(data['Club'])):
        salaries.loc[n] = data.loc[z]
        n+=1

salaries
```
