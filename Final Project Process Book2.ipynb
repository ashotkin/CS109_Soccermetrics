```python
## This code is for scraping the MLS site for the results of games and storing them in a dataframe
##### DON'T RUN

resultpage = "http://www.mlssoccer.com/results/"


results = pd.DataFrame(columns=('Year', 'Club', 'Date', 'Versus', 'Club Score', 'Opposing Score', 'W/L/D'))
n=0
for year in range(2011,2015):
    page = BeautifulSoup(urllib2.urlopen(resultpage + str(year)).read(), "html.parser")
    table = page.find('table')
    for row in table.findAll('tr')[1:]:
        col = row.findAll('td')
        for i in range(len(col)):
            if i == 0:
                team = col[i]
                clubsplit = str(team).split('\n')
                club = clubsplit[2][10:-16]
            else:
                data = str(col[i]).split('>')
                winloss = data[2][0]
                date =  data[3][:-4]
                versussplit = data[4].split()
                versus = ' '
                for j in range(len(versussplit)):
                    versus = versus + ' ' + versussplit[j]
                versus = versus[5:-4]
                split = data[5].split()
                score1 = int(split[0])
                score2 = int(split[2])
                

                if winloss == 'W':
                    if score1<score2:
                        score = score2
                        scoreopp = score1
                    else:
                        score = score1
                        scoreopp = score2
                elif winloss == 'L':
                    if score1<score2:
                        score = score1
                        scoreopp = score2
                    else:
                        score = score2
                        scoreopp = score1
                else:
                    score = score1
                    scoreopp = score2
                    
                results.loc[n] = [year, club, date, versus, score, scoreopp, winloss]
                n+=1

            
results
```

```python
## This code just goes through and assigns a club to each player for each game.

#This frame (if run) prints out the index every 10000 iterations, so it's possible
#to track how far along the code is. 
fails = []
positions = []

for index, row in player_games.iterrows():
    player2,year = player_games.ix[index][['Name','Year']]
    sliceOfPlayers = players[players['Player'] == player2]
    sliceOfPlayers = sliceOfPlayers[sliceOfPlayers['Year'] == str(year)] 
    club = sliceOfPlayers['Club'].tolist()
    if club:
        player_games.loc[index,['Club']] = sliceOfPlayers['Club'].tolist()[0]
        positions.append(sliceOfPlayers['Position'].tolist()[0])
    else:
        player_games.loc[index,['Club']] = 'failure'
        positions.append('failure')
        fails.append(index)
    if index % 10000 == 0:
        print index
player_games['POS'] = positions
```

```python
## Run this to create the game dictionary
## We chose to only include offensive players because of the dearth of publicly available
## stats on defensive players (such stats ideally would have included information on passing,
## challenges, slide tackles, touches etc)
offense = ['F', 'F-M', 'M', 'M-F']
temp = player_games.drop_duplicates()
temp = temp[temp['POS'].isin(offense)]
temp = temp.groupby(by =['Club','Date','Name'], as_index=False).sum()
teams = set(temp['Club'])
games = {}
for team in teams:
    game = {}
    temp2 = temp[temp['Club'] == team]
    dates = set(temp2['Date'])
    for date in dates:
        players1 = {}
        tempPlayers = temp2[temp2['Date'] == date]
        for player in set(tempPlayers['Name']):
            players1[player] = (tempPlayers[tempPlayers['Name'] == player])['Minutes'].tolist()[0]
        game[date] = players1
    games[team] = game
```

```python
## creates keepers dataframe
keepers = player_games.drop_duplicates()
keepers = keepers[keepers['POS'] == 'GK']
keepers = keepers.groupby(by =['Club','Date','Name'], as_index=False).sum()
teams = set(temp['Club'])
goalies = {}
for team in teams:
    goalie = {}
    keepers2 = keepers[keepers['Club'] == team]
    dates = set(keepers2['Date'])
    for date in dates:
        guys = {}
        tempGoalies = keepers2[keepers2['Date'] == date]
        for player1 in set(tempGoalies['Name']):
            guys[player1] = (tempGoalies[tempGoalies['Name'] == player1])['Minutes'].tolist()[0]
        goalie[date] = guys
    goalies[team] = goalie
```

```python
## this code normalizes all of the player statistics by how many minutes the player played

players = players[players['MINS'] != 0]
players = players[players['SOG'] != '0/0']

finalResults = pd.DataFrame()
people = players['Player']
years = players['Year']
assists = []
shots = []
shotsOnGoal = []
gameWinningGoals = []
pkAverage = []
homeGoals = []
roadGoals = []

for index, row in players.iterrows():
    minutes = float(row[6])/90
    if minutes == 0:
        print row[0]
    assists.append(row[8] / minutes)
    shots.append(row[9]/ minutes)
    shotsOnGoal.append(int(row[10])/ minutes)
    gameWinningGoals.append(row[11]/minutes)
    percents = row[12].split('/')
    if int(percents[1]) == 0:
        pkAverage.append(0)
    else:
        pkAverage.append(int(percents[0])/float(percents[1]))
    homeGoals.append(row[13]/minutes)
    roadGoals.append(row[14]/minutes)
    
```

```python
## this returns the average player stats normalized by minutes for a certain team in a certain game

def calculator (year, dictionary):
    totalMinutes = float(0)
    totals = [0] * 7
    rows = []
    for player in dictionary.keys():
        averageGuy = averagedPlayers[averagedPlayers['Player'] == player]
        averageGuy = averageGuy[averageGuy['Year'] == str(year)]
        if not(averageGuy.empty):
            individualStats = dictionary[player] * averageGuy[['A/90min', 'GWG/90min', 'HmG/90min', 'PKG/A', 'RdG', 'SHTS/90min', 'SOG/90min']]
            if individualStats.shape[0] > 1:
                print player,year
            
            totals = np.add(totals, individualStats)
            totalMinutes += dictionary[player]
 
    return (totals/totalMinutes).values
```

```python
## Calculates minutes-weighted average of goalkeeper stats for a given game

def calculatorGoalie (year, dictionary):
    totalMinutes = float(0)
    totals = [0] * 3
    rows = []
    for player in dictionary.keys():
        averageGuy = goalkeepers[goalkeepers['Player'] == player]
        averageGuy = averageGuy[averageGuy['Year'] == year]
        if not(averageGuy.empty):
            individualStats = dictionary[player] * averageGuy[['SHTS', 'SV', 'GA']]
            if individualStats.shape[0] > 1:
                print player,year
            
            totals = np.add(totals, individualStats / averageGuy['MINS'].tolist()[0])
            totalMinutes += dictionary[player]
 
    return (totals/totalMinutes).values
```
