## Scraping the data into CSVs

We ran the code in this section in order to scrape our data, and now that we've done that, we saved it into CSV's so we never have to scrape it again. The CSV's are available above.

As for collecting our data, we scraped the mlssoccer.com website for every player, team, and game statistic available. Conveniently, all of their data was written into the html, so we only had to scrape the html. Scraping the information on individual players’ contributions to each game was the most tedious part of scraping mlssoccer.com: It involved loading a page for every single player active in the MLS since 2011, scraping that page for every game that player has played in and how many minutes he played in that game, associating the player with his club and storing that information and the date of the game in a dataframe. This took up to 10 seconds per player, and given there were ~2500 players in the database this added up to several hours. We added a try-except block so that we could catch players that didn’t have stat pages, deal with each player and then continue the gathering without redoing the entire set of players. Including these delays, the process took about 7 hours. As for the salary information, this information was stored on an outside website in the form of a PDF, and we had to employ new techniques in order to convert this data into files that could be read into our iPython notebook without having to rely on modules that approximated PDF locations. We used Adobe Acrobat to convert the PDF’s into CSV files, which we had to open and adjust the spacing and format so that they could be read into a DataFrame using PANDAS. Reformatting the files was the most difficult part of the process because each file was exported differently, so we could not apply a universal reformatting algorithm to every file and instead had to do each one by hand. While the converting and formatting were tedious, as soon as we had the CSV’s, importing the data was straightforward.

```python
## This cell scrapes player season statistics for creating initial player profiles
###### NEVER RUN THIS CELL AGAIN ######

# I didn't know how to scrape before. So I used this webpage as a reference:
# http://palewi.re/posts/2008/04/20/python-recipe-grab-a-page-scrape-a-table-download-a-file/

mainpage = "http://www.mlssoccer.com/stats/season"

# the table is continued on other pages
#urls = [mainpage + '?page={}'.format(i) for i in range(1,26)]
#urls.insert(0,mainpage)

playerpage = "http://www.mlssoccer.com/stats/season?"
players = pd.DataFrame(columns=('Player', 'Year', 'Club', 'Position', 'GP', 'GS', 'MINS', 'G', 'A', 'SHTS', 'SOG', 'GWG', 'PKG/A', 'HmG', 'RdG', 'G/90min', 'SC%'))
i=0
for year in range(1996,2015):
    pageNumber = 0
    last_page = 0
    while pageNumber <= last_page:
        page = BeautifulSoup(urllib2.urlopen(playerpage + "season_year=" + str(year) + "&page=" + str(pageNumber)).read(), 'html.parser')
        if pageNumber == 0:
            last_page = int(str(page.find('li', {"class":"pager-last last"}).contents[0]).split('=')[3].split('&')[0])
        pageNumber+=1
        table = page.find('table')
        for row in table.findAll('tr')[1:]:
            col = row.findAll('td')
            player = col[0].string
            club = col[1].string
            pos = col[2].string
            gp = int(col[3].string)
            gs = int(col[4].string)
            mins = int(col[5].string)
            g = int(col[6].string)
            a = int(col[7].string)
            shts = int(col[8].string)
            sog = int(col[9].string)
            gwg = int(col[10].string)
            pkga = col[11].string
            hmg = int(col[12].string)
            rdg = int(col[13].string)
            g90min = float(col[14].string)
            sc = float(col[15].string)
            players.loc[i] = [player, year, club, pos, gp, gs, mins, g, a, shts, sog, gwg, pkga, hmg, rdg, g90min, sc]
            i+=1
            
players
```

```python
## This code is for scraping the MLS site for the results of games and storing them in a dataframe
##### DON'T RUN

resultpage = "http://www.mlssoccer.com/results/"


results = pd.DataFrame(columns=('Year', 'Club', 'Date', 'Versus', 'Club Score', 'Opposing Score', 'W/L/D'))
n=0
for year in range(2011,2015):
    page = BeautifulSoup(urllib2.urlopen(resultpage + str(year)).read(), "html.parser")
    table = page.find('table')
    for row in table.findAll('tr')[1:]:
        col = row.findAll('td')
        for i in range(len(col)):
            if i == 0:
                team = col[i]
                clubsplit = str(team).split('\n')
                club = clubsplit[2][10:-16]
            else:
                data = str(col[i]).split('>')
                winloss = data[2][0]
                date =  data[3][:-4]
                versussplit = data[4].split()
                versus = ' '
                for j in range(len(versussplit)):
                    versus = versus + ' ' + versussplit[j]
                versus = versus[5:-4]
                split = data[5].split()
                score1 = int(split[0])
                score2 = int(split[2])
                

                if winloss == 'W':
                    if score1<score2:
                        score = score2
                        scoreopp = score1
                    else:
                        score = score1
                        scoreopp = score2
                elif winloss == 'L':
                    if score1<score2:
                        score = score1
                        scoreopp = score2
                    else:
                        score = score2
                        scoreopp = score1
                else:
                    score = score1
                    scoreopp = score2
                    
                results.loc[n] = [year, club, date, versus, score, scoreopp, winloss]
                n+=1

            
results
```

```python
## Run this to create the game dictionary
## We chose to only include offensive players because of the dearth of publicly available
## stats on defensive players (such stats ideally would have included information on passing,
## challenges, slide tackles, touches etc)
offense = ['F', 'F-M', 'M', 'M-F']
temp = player_games.drop_duplicates()
temp = temp[temp['POS'].isin(offense)]
temp = temp.groupby(by =['Club','Date','Name'], as_index=False).sum()
teams = set(temp['Club'])
games = {}
for team in teams:
    game = {}
    temp2 = temp[temp['Club'] == team]
    dates = set(temp2['Date'])
    for date in dates:
        players1 = {}
        tempPlayers = temp2[temp2['Date'] == date]
        for player in set(tempPlayers['Name']):
            players1[player] = (tempPlayers[tempPlayers['Name'] == player])['Minutes'].tolist()[0]
        game[date] = players1
    games[team] = game
```

```python
## this returns the average player stats normalized by minutes for a certain team in a certain game

def calculator (year, dictionary):
    totalMinutes = float(0)
    totals = [0] * 7
    rows = []
    for player in dictionary.keys():
        averageGuy = averagedPlayers[averagedPlayers['Player'] == player]
        averageGuy = averageGuy[averageGuy['Year'] == str(year)]
        if not(averageGuy.empty):
            individualStats = dictionary[player] * averageGuy[['A/90min', 'GWG/90min', 'HmG/90min', 'PKG/A', 'RdG', 'SHTS/90min', 'SOG/90min']]
            if individualStats.shape[0] > 1:
                print player,year
            
            totals = np.add(totals, individualStats)
            totalMinutes += dictionary[player]
 
    return (totals/totalMinutes).values
```

```python
## Calculates minutes-weighted average of goalkeeper stats for a given game

def calculatorGoalie (year, dictionary):
    totalMinutes = float(0)
    totals = [0] * 3
    rows = []
    for player in dictionary.keys():
        averageGuy = goalkeepers[goalkeepers['Player'] == player]
        averageGuy = averageGuy[averageGuy['Year'] == year]
        if not(averageGuy.empty):
            individualStats = dictionary[player] * averageGuy[['SHTS', 'SV', 'GA']]
            if individualStats.shape[0] > 1:
                print player,year
            
            totals = np.add(totals, individualStats / averageGuy['MINS'].tolist()[0])
            totalMinutes += dictionary[player]
 
    return (totals/totalMinutes).values
```
